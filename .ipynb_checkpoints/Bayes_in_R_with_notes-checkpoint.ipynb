### Setting up the A/B test
Building the Prior, let's assume that: 

-600 users, divided into 2 groups á 300

-clickthrough rate of about 30% -> no strong belief so beta(3,7) fits good into our beliefs

```r
alpha_prior = 3
beta_prior = 7
xs <- seq(0,1,by=0.01)
plot(xs,dbeta(xs,alpha_prior,beta_prior),type='l',lwd=3,
     ylab="density",
     xlab="probability of clickthrough",
     main="PDF Beta(3,7)")
```

Beta (α posterior, β posterior ) = Beta (α prior + α likelihood, β prior + β likelihood )

```r
# neue Daten kommen rein:
alpha_likelihood_A = 36
beta_likelihood_A = 114
alpha_likelihood_B = 50
beta_likelihood_B = 100

plot(xs,dbeta(xs,alpha_prior + alpha_likelihood_A,beta_prior + beta_likelihood_A),col="red",type='l',lwd=3,
     ylab="density",
     xlab="probability of clickthrough",
     main="PDF Beta()")
lines(xs,dbeta(xs,alpha_prior + alpha_likelihood_B, beta_prior + beta_likelihood_B), col="blue",lwd=3)

legend(x = "topright",          # Position
       legend = c("var A", "var B"),  # Legend texts
       lty = c(1, 1),           # Line types
       col = c(2, 4),           # Line colors
       lwd = 3)                 # Line width
```

```r
# Variante B scheint erfolgreicher zu sein.
# MC Sampling um die Hypothese zu bestätigen:
```

Random Sampling mit Monte Carlo simulation:

```r
# 100000 Versuche:
n.trials <- 100000

#Samples aus den 10000 Versuchen mit Posterior-Verteilung
a.samples <- rbeta(n.trials, alpha_prior + alpha_likelihood_A, beta_prior + beta_likelihood_A)
b.samples <- rbeta(n.trials, alpha_prior + alpha_likelihood_B, beta_prior + beta_likelihood_B)

# in wievielen Tests von allen Tests sind die B-Samples besser als die A-Samples?
p.b_superior <- sum(b.samples > a.samples)/n.trials
p.b_superior
```

~ 96% Chance, dass Variante B besser performt. Aber wie viel besser genau?

```r
ratio = b.samples/a.samples
```

```r
options(repr.plot.width=8, repr.plot.height=4)
hist(ratio)
```

```r
plot(ecdf(ratio))
```
