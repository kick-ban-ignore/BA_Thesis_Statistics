```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pymc3 as pm
```

```python
# Generate prior draws assuming avg of 15% -> beta of 10/50 fits the prior belief

alpha = 18.25
beta = 6.75
kappa = 25
clicks = 17
trials = 20

prior_draws = np.random.beta(alpha, beta, 1000)

# Plot the prior
sns.kdeplot(prior_draws, shade=True, label="prior")
plt.show()
```

```python
pm.plot_posterior(prior_draws, hdi_prob=0.95)
plt.show()
```

```python
import statistics as st
print(prior_draws.mean(), st.median(prior_draws), st.mode(prior_draws))
```

```python
#Plot Likelihood
likelihood = np.random.beta(clicks, trials-clicks, 1000)

# Plot the prior
sns.kdeplot(likelihood, shade=True, label="likelihood")
plt.show()
```

```python
print(likelihood.mean(), st.median(likelihood))
```

```python
# Generate Posteriors

def simulate_beta_posterior(clicks, trials, beta_prior_a, beta_prior_b):  
    posterior_draws = np.random.beta(clicks + beta_prior_a,trials - clicks + beta_prior_b, 1000)
    
    return posterior_draws
```

```python
# Simulate posterior draws for each product
posterior = simulate_beta_posterior(clicks, trials, alpha, beta)
```

```python
sns.kdeplot(posterior, shade=True, label="clothes")
plt.show()
```

```python
print(posterior.mean(), st.median(posterior), st.mode(posterior))
```

```python
pm.plot_posterior(posterior, hdi_prob=0.95)
plt.show()
```
